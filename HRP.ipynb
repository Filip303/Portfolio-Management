```python
import yfinance as yf
import pandas as pd
from pypfopt import HRPOpt, risk_models, expected_returns
import matplotlib.pyplot as plt

tickers = ['BTC-USD','O','MVUS.MI','EGLN.L','MSTR']
data = yf.download(tickers, period='10y')['Adj Close']

returns = data.pct_change().dropna()

cov_matrix = risk_models.sample_cov(returns)

hrp = HRPOpt(returns)
weights = hrp.optimize()

print('Pesos de la cartera HRP:')
print(weights)

pd.Series(weights).plot.pie(figsize=(10,10), autopct='%1.1f%%')
plt.title('Composici√≥n de la cartera optimizada por HRP')
plt.show()
```

```python
import yfinance as yf
import quantstats as qs
import pandas as pd

# Paso 1: Descargar los datos de los activos
tickers = ['URTH', 'BTC-USD','MSTR','O','MVUS.MI','EGLN.L']
data = yf.download(tickers, period='10y')['Adj Close']

# Asegurarse de que todas las fechas sean tz-naive (sin zona horaria)
data.index = data.index.tz_localize(None)

# Paso 2: Calcular los retornos diarios
returns = data.pct_change().dropna()

# Paso 3: Seleccionar URTH como benchmark y calcular los retornos de la cartera √≥ptima
# Cambiar de un set a un dict para optimal_weights
optimal_weights = {'BTC-USD': 0.02151584148431003,
                   'EGLN.L': 0.49778251791865297,
                   'MSTR': 0.021020556189659742,
                   'MVUS.MI': 0.3469487629916409,
                   'O': 0.11273232141573637}

# Filtrar los retornos de los activos en la cartera y calcular los retornos ponderados
selected_returns = returns[list(optimal_weights.keys())]
portfolio_returns = (selected_returns * pd.Series(optimal_weights)).sum(axis=1)

# Paso 4: Definir la tasa libre de riesgo anual y convertirla a diaria
risk_free_rate_annual = 0.025
risk_free_rate_daily = risk_free_rate_annual / 252

# Generar el informe completo con QuantStats
qs.reports.full(portfolio_returns, benchmark=returns['URTH'], rf=risk_free_rate_daily)
```
